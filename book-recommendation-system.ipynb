## Collaborative filtering based recommendation system

```python
rating_user = ratings_books.groupby("User-ID").count()["Book-Rating"] >200
rating_index_user = rating_user[rating_user].index
rating_index_user
```

```python
ratings_books["User-ID"]
```

```python
famous_books
```

```python
pivited_table = famous_books.pivot_table( index = "Book-Title",columns = "User-ID",values = "Book-Rating")
pivited_table.fillna(0,inplace = True)
pivited_table
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pivited_table)
```

```python
pickle.dump(similarity_score,open("similarity_collab.pkl","wb"))
```

```python
pickle.dump(pivited_table,open("pivited_table.pkl","wb"))
```

```python
np.where(pivited_table.index == "Fahrenheit 451")[0][0]
```

```python
np.where(pivited_table.index=="Vector")
```

```python
def recommend_user(book):
    index = np.where(pivited_table.index==book)[0][0]
    filtered_books = sorted(list(enumerate(similarity_score[index])),key = lambda x: x[1],reverse = True)[1:6]
    collaborative_list = []
    for i in filtered_books:
        collaborative_list.append(pivited_table.index[(i[0])])
    return collaborative_list
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
```

```python
def stems(tags):
    y= []
    for i in tags.split():
        y.append(stemmer.stem(i))
    return " ".join(y)
```

```python
books_with_tags["Tags"] = books_with_tags["Tags"].apply(stems)
```

```python
books_with_tags
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommendation_content(book):
    index = books_with_tags[books_with_tags["Book-Title"]==book].index[0]
    distances = similarity[index]
    movies_list= sorted(list(enumerate(distances)),key = lambda x:x[1], reverse = True)[1:6]
    content_list = []
    for i in movies_list:
        content_list.append(books_with_tags.iloc[i[0]]["Book-Title"])
    return content_list
    
```

# Recommendation System based on Hybrid Approach

```python
def hybrid_recommendation(book):
    hybrid = recommendation_content(book) + recommend_user(book)
    hybrid_based = set(hybrid)
    strin =  '\n'.join(hybrid_based)
    return strin
```

```python
result = hybrid_recommendation("The Notebook")
print(result)
```
